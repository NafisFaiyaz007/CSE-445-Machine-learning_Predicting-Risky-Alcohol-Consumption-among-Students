```python
# Create a DataFrame with a categorical column
data = pd.DataFrame({'Color': ['Red', 'Green', 'Blue', 'Green']})

# Create an instance of the encoder
encoder = OneHotEncoder()

# Fit and transform the categorical column
encoded_data = encoder.fit_transform(data[['Color']]).toarray()

# Replace the original column with the encoded features
data_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Color']))
data = data.drop(columns=['Color'])
data = pd.concat([data, data_encoded], axis=1)
```
